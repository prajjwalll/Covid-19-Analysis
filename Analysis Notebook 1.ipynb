## Grouping Data Country wise

```python
confirmed_aggregated = corona_dataset_confirmed.groupby('Country/Region').sum()
confirmed_aggregated.head()
```

```python
deaths_aggregated = corona_dataset_deaths.groupby('Country/Region').sum()
deaths_aggregated.head()
```

```python
recovered_aggregated = corona_dataset_recovered.groupby('Country/Region').sum()
recovered_aggregated.head()
```

```python
list1 = [confirmed_aggregated,deaths_aggregated,recovered_aggregated]
for i in list1:
    print(i.info())
    print('-'*45)
```

There are total 188 distinct countries.

## Calculating the Infection Rate of Countries

```python
countries = confirmed_aggregated.index
inf_rate = []
for i in countries:
    inf_rate.append(confirmed_aggregated.loc[i].diff().max())
confirmed_aggregated['Infection Rate'] = inf_rate
```

```python
confirmed_data = pd.DataFrame(confirmed_aggregated['Infection Rate'].sort_values(ascending=False))
confirmed_data
```

## Calculating the Death rate of countries

```python
countries = deaths_aggregated.index
deaths_rate = []
for i in countries:
    deaths_rate.append(deaths_aggregated.loc[i].diff().max())
deaths_aggregated['Death Rate'] = deaths_rate
```

```python
deaths_data = pd.DataFrame(deaths_aggregated['Death Rate'].sort_values(ascending=False))
deaths_data
```

## Calculating Recovery rate of the countries

```python
countries = recovered_aggregated.index
recovery_rate = []
for i in countries:
    recovery_rate.append(recovered_aggregated.loc[i].diff().max())
recovered_aggregated['Recovery Rate'] = recovery_rate
```

```python
recovered_data = pd.DataFrame(recovered_aggregated['Recovery Rate'].sort_values(ascending=False))
recovered_data
```

## Merging all data 

```python
temp = confirmed_data.join(deaths_data).copy()
final_data = temp.join(recovered_data).copy()
final_data.head()
```

```python
whr_dataset.drop(['Overall rank'],axis=1,inplace=True)
whr_dataset.head()
```

```python
whr_dataset.rename(columns={'Country or region':'Country/Region'},inplace=True)
whr_dataset.head()
```

```python
whr_dataset.set_index('Country/Region',inplace=True)
```

```python
whr_dataset
```

```python
whr_dataset.shape
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x=whr_dataset.head(10).index,y=whr_dataset.head(10).Score)
plt.xticks(rotation='45')
plt.title('Countries vs World Happiness Score out of 10')
plt.show()
```

## Merging World Happiness Report with Corona Final Data

```python
total_data = whr_dataset.join(final_data).copy()
total_data.head()
```

```python
# sorting values with respect to Infection Rate of the countries.
total_data.sort_values(by=['Infection Rate'],inplace=True,ascending=False)
total_data.head()
```

```python
# checking country data
total_data.loc['India']
```

## Analysis on the basis of World Happiness Report 2019

```python
total_data.head()
```

```python
# calculating the correlation between the all the attributes
matrix = total_data.corr()
matrix
```

```python
# visualize the correlation
plt.figure(figsize=(8,8))
sns.heatmap(matrix, vmin = -1, vmax = 1, center = 0, square = True, linewidths = 0.05,cmap = "YlGnBu")
plt.show()
```

```python
# visualizing GDP per capita vs Infection Rate
x = total_data['GDP per capita']
y = total_data['Infection Rate']
sns.scatterplot(x,np.log(y))   
plt.show()
# Best fit line to show the relation.
sns.regplot(x,np.log(y))
plt.show()
```

```python
# visualizing Social Support vs Infection Rate
x = total_data['Social support']
y = total_data['Infection Rate']
sns.scatterplot(x,np.log(y))   
plt.show()
# Best fit line to show the relation.
sns.regplot(x,np.log(y))
plt.show()
```

```python
# visualizing Healthy life expectancy vs Infection Rate
x = total_data['Healthy life expectancy']
y = total_data['Infection Rate']
sns.scatterplot(x,np.log(y))   
plt.show()
# Best fit line to show the relation.
sns.regplot(x,np.log(y))
plt.show()
```

## Conclusion
1. The countries shows linear relation with GDP per capita, Social Support and Healthy life expectancy.

2. It implies that the countries which have higher GDP per capita have higher infection rate.

3. The countries which have higher social support have higher infection rate.

4. The countries which have higher healthy life expectancy have higher infection rate.
